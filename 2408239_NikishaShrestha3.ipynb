```python
#Which adult male passenger (age >= 18 and Sex is 'male') paid the highest fare relative to their class?
#Question1
from google.colab import files
import pandas as pd
import io
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/AIWeek2/Titanic-Dataset.csv")
df['fare_per_class'] = df['Fare'] / df['Pclass']
print(df.head())
```

```python
#3.3
#What percent of the total fare revenue came from each passenger class?
#Question1
total_fare = df['Fare'].sum()
print(total_fare)
```

```python
#Question2
first_class_passengers = df[df['Pclass'] == 1]
first_class_fare = first_class_passengers['Fare'].sum()
print(first_class_fare)
```

```python
#Question3
second_class_passengers = df[df['Pclass'] == 2]
second_class_fare = second_class_passengers['Fare'].sum()
print("Second class fare:", second_class_fare)
third_class_passengers = df[df['Pclass'] == 3]
third_class_fare = third_class_passengers['Fare'].sum()
print("Third class fare:", third_class_fare)
```

```python
#Question4
fare_totals = [first_class_fare, second_class_fare, third_class_fare]
print(f"Fare totals for each class: {fare_totals}")
```

```python
#Question5
total_fare = df['Fare'].sum()
first_class_proportion = first_class_fare / total_fare
second_class_proportion = second_class_fare / total_fare
third_class_proportion = third_class_fare / total_fare
fare_proportions = [first_class_proportion, second_class_proportion, third_class_proportion]
print(f"Fare proportions for each class: {fare_proportions}")
```

```python
#What percent of the total number of passengers on the Titanic belonged to each age group (e.g., child, adult, senior)?
#Question1
df['age_group'] = pd.cut(df['Age'], bins=[0, 17, 64, float('inf')], labels=['child', 'adult', 'senior'])
print(df[['Age', 'age_group']].head())
```

```python
#Question2
total_passengers = df.shape[0]
print(f"Total number of passengers: {total_passengers}")
```

```python
#Question3
age_group_counts = df['age_group'].value_counts()
age_group_proportions = age_group_counts / total_passengers
print(age_group_proportions)
```

```python
#Question4
age_group_percentages = age_group_proportions * 100
print(age_group_percentages)
```
